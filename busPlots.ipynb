```python
import plotly.plotly as py
import pandas as pd

py.sign_in('orgiele', '0hLF1CVCaH5sEU0CG9Ah')

from plotly.graph_objs import *

mapbox_access_token = 'pk.eyJ1Ijoib3JnaWVsZSIsImEiOiJjajJ4azAyYnkwMHZ4MnFxbm41d3JocHlzIn0.GDMQKR_XxQIZtPX8dq6K4g'

nightBuses14 = pd.concat([pd.read_json('file:data/14-05/buses-night-filtered.json'),
                       pd.read_json('file:data/14-05/buses2-night-filtered.json')])

normalBuses14 = pd.concat([pd.read_json('file:data/14-05/buses-normal-filtered.json'),
                        pd.read_json('file:data/14-05/buses2-normal-filtered.json'),
                        pd.read_json('file:data/14-05/buses3-normal-filtered.json')])

zoneBuses14 = pd.concat([pd.read_json('file:data/14-05/buses-zone-filtered.json'),
                        pd.read_json('file:data/14-05/buses2-zone-filtered.json'),
                        pd.read_json('file:data/14-05/buses3-zone-filtered.json')])

zoneSupplementaryBuses14 = pd.concat([pd.read_json('file:data/14-05/buses2-zone-supplementary-filtered.json'),
                       pd.read_json('file:data/14-05/buses3-zone-supplementary-filtered.json')])

zonePeriodicBuses14 = pd.read_json('file:data/14-05/buses3-zone-periodic-filtered.json')

fastBuses14 = pd.concat([pd.read_json('file:data/14-05/buses-fast-filtered.json'),
                        pd.read_json('file:data/14-05/buses2-fast-filtered.json'),
                        pd.read_json('file:data/14-05/buses3-fast-filtered.json')])

fastPeriodicBuses14 = pd.concat([pd.read_json('file:data/14-05/buses2-fast-periodic-filtered.json'),
                        pd.read_json('file:data/14-05/buses3-fast-periodic-filtered.json')])

def plotBuses(df, colors, file, pitch, zoom):
    data = Data([
        Scattermapbox(
            lat=df['Lat'],
            lon=df['Lon'],
            mode='markers',
            hoverinfo='text',
            marker=Marker(
                size=3,
                opacity=0.7,
                color=colors,
                colorscale='Portland' 
                #Pre-defined color scales - 'pairs' | 'Greys' | 'Greens' | 'Bluered' | 'Hot' | 'Picnic' | 
                #'Portland' | 'Jet' | 'RdBu' | 'Blackbody' | 'Earth' | 'Electric' | 'YIOrRd' | 'YIGnBu' 
            ),
            text=df['Lines']
        )
    ])

    layout = Layout(
        autosize=True,
        hovermode='closest',
        mapbox=dict(
            accesstoken=mapbox_access_token,
            bearing=0,
            center=dict(
                lat=52.228626,
                lon=21.005654
            ),
            pitch=pitch,
            zoom=zoom
        ),
    )

    fig = dict(data=data, layout=layout)
    py.image.save_as(fig, filename=file)
    from IPython.display import Image
    return Image(file)
```
